```python
# %% [markdown]
# # TP Deep Learning - Partie 1: Analyse des données
# ## Bank Telemarketing Dataset
# 
# **Auteur:** [Votre Nom]
# **Date:** 2025-2026

# %%
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import warnings
warnings.filterwarnings('ignore')

# Configuration
plt.style.use('default')
sns.set_palette("husl")
plt.rcParams['figure.figsize'] = (12, 6)

print("=" * 60)
print("ANALYSE DES DONNÉES - BANK TELEMARKETING")
print("=" * 60)

# %%
# Chargement des données
data_path = '../datasets/bank_additional/bank-additional.csv'

if not os.path.exists(data_path):
    print(f"❌ Fichier non trouvé: {data_path}")
    print("Exécutez d'abord: python download_data.py")
else:
    df = pd.read_csv(data_path, sep=';')
    print(f"✅ Données chargées avec succès!")
    print(f"Shape: {df.shape}")

# %%
# ============================================================
# 1. INFORMATIONS GÉNÉRALES
# ============================================================
print("\n" + "=" * 60)
print("1. INFORMATIONS GÉNÉRALES")
print("=" * 60)

print(f"📊 Shape du dataframe: {df.shape}")
print(f"📝 Nombre d'instances: {df.shape[0]:,}")
print(f"📝 Nombre de caractéristiques: {df.shape[1]}")
print(f"\n🎯 Variable cible: 'y'")
print(f"   → Objectif: Prédire si le client souscrit à un dépôt bancaire (yes/no)")

# %%
# ============================================================
# 2. TYPES DE VARIABLES
# ============================================================
print("\n" + "=" * 60)
print("2. TYPES DE VARIABLES")
print("=" * 60)

numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
categorical_cols = df.select_dtypes(include=['object']).columns.tolist()
if 'y' in categorical_cols:
    categorical_cols.remove('y')

print(f"🔢 Variables numériques ({len(numeric_cols)}): {numeric_cols}")
print(f"🏷️ Variables catégorielles ({len(categorical_cols)}): {categorical_cols}")
print(f"\n💡 Type d'apprentissage: Classification binaire supervisée")

# %%
# ============================================================
# 3. DISTRIBUTION DES CLASSES
# ============================================================
print("\n" + "=" * 60)
print("3. DISTRIBUTION DES CLASSES")
print("=" * 60)

class_counts = df['y'].value_counts()
print(class_counts)

# Pourcentages
yes_pct = class_counts['yes']/len(df)*100
no_pct = class_counts['no']/len(df)*100
print(f"\n📊 Proportion 'yes': {yes_pct:.2f}%")
print(f"📊 Proportion 'no': {no_pct:.2f}%")

# Visualisation
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Bar plot
bars = class_counts.plot(kind='bar', ax=axes[0], color=['#ff6b6b', '#4ecdc4'])
axes[0].set_title('Distribution des classes', fontsize=12, fontweight='bold')
axes[0].set_xlabel('Souscription', fontsize=10)
axes[0].set_ylabel('Nombre', fontsize=10)
axes[0].set_xticklabels(['Non (no)', 'Oui (yes)'], rotation=0)

# Ajout des valeurs sur les barres
for i, bar in enumerate(bars.patches):
    axes[0].text(bar.get_x() + bar.get_width()/2, bar.get_height() + 500,
                f'{int(bar.get_height()):,}', ha='center', va='bottom')

# Pie chart
colors = ['#ff6b6b', '#4ecdc4']
axes[1].pie(class_counts.values, labels=['Non (no)', 'Oui (yes)'], 
            autopct='%1.1f%%', colors=colors, startangle=90, explode=(0.05, 0))
axes[1].set_title('Proportion des classes', fontsize=12, fontweight='bold')

plt.suptitle("Analyse de la variable cible", fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('../reports/figures/class_distribution.png', dpi=100, bbox_inches='tight')
plt.show()

# %%
# ============================================================
# 4. DISCUSSION DES MÉTRIQUES DE PERFORMANCE
# ============================================================
print("\n" + "=" * 60)
print("4. DISCUSSION DES MÉTRIQUES DE PERFORMANCE")
print("=" * 60)

print(f"""
📈 Analyse du déséquilibre des classes:
   • Classe majoritaire (no): {no_pct:.1f}% des données
   • Classe minoritaire (yes): {yes_pct:.1f}% des données

⚠️ Problème potentiel:
   Un modèle qui prédit toujours "no" aurait une accuracy de {no_pct:.1f}%
   mais serait INUTILE car il ne détecterait AUCUN client potentiel.

✅ Métriques recommandées pour ce problème:

   1. F1-Score (le plus important):
      - Moyenne harmonique entre Precision et Recall
      - Idéal pour données déséquilibrées
      - Formule: F1 = 2 × (Precision × Recall) / (Precision + Recall)

   2. Precision:
      - "Parmi les clients prédits positifs, combien le sont vraiment ?"
      - Important pour éviter de contacter trop de faux positifs

   3. Recall (Sensibilité):
      - "Parmi les vrais clients positifs, combien ai-je détectés ?"
      - Critique pour ne pas rater des opportunités

   4. AUC-ROC:
      - Performance globale indépendante du seuil
      - Idéal pour comparer les modèles

   5. Courbe ROC:
      - Visualise le compromis TPR (Recall) vs FPR
      
⚠️ À ÉVITER: L'accuracy seule (trompeuse sur données déséquilibrées)
""")

# %%
# ============================================================
# 5. STATISTIQUES DESCRIPTIVES
# ============================================================
print("\n" + "=" * 60)
print("5. STATISTIQUES DESCRIPTIVES")
print("=" * 60)

display(df[numeric_cols].describe())

# %%
# ============================================================
# 6. ORGANISATION DES INSTANCES
# ============================================================
print("\n" + "=" * 60)
print("6. ORGANISATION DES INSTANCES")
print("=" * 60)

print("📋 Organisation des données:")
print("   • Pas d'ordre temporel spécifique dans le fichier")
print("   • Les instances sont mélangées aléatoirement")
print("   • 20 variables descriptives (10 numériques, 10 catégorielles)")
print("   • 1 variable cible binaire (yes/no)")
print("   • Données issues de campagnes de télémarketing (2008-2013)")
print("\n📊 Structure:")
print("   • Lignes: chaque client contacté")
print("   • Colonnes: attributs du client + contexte économique")
print("   • Target: souscription (yes) ou non (no)")

# %%
# ============================================================
# 7. MATRICE DE CORRÉLATION
# ============================================================
print("\n" + "=" * 60)
print("7. MATRICE DE CORRÉLATION")
print("=" * 60)

# Encodage de la cible pour corrélation
df_encoded = df.copy()
df_encoded['y_encoded'] = df_encoded['y'].map({'no': 0, 'yes': 1})

if len(numeric_cols) > 0:
    corr_matrix = df_encoded[numeric_cols + ['y_encoded']].corr()
    corr_with_target = corr_matrix['y_encoded'].sort_values(ascending=False)
    
    print("\n🔗 Corrélations avec la variable cible (souscription):")
    print("   (plus proche de +1 ou -1 = plus influent)")
    print("-" * 50)
    for var, corr in corr_with_target.items():
        if var != 'y_encoded':
            emoji = "🔴" if abs(corr) > 0.3 else "🟡" if abs(corr) > 0.1 else "🟢"
            print(f"   {emoji} {var:20s}: {corr:+.4f}")
    
    # Visualisation
    plt.figure(figsize=(12, 10))
    mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
    sns.heatmap(corr_matrix, mask=mask, annot=True, fmt='.2f', 
                cmap='coolwarm', center=0, square=True,
                linewidths=0.5, cbar_kws={"shrink": 0.8})
    plt.title('Matrice de corrélations', fontsize=14, fontweight='bold')
    plt.tight_layout()
    plt.savefig('../reports/figures/correlation_matrix.png', dpi=100, bbox_inches='tight')
    plt.show()
    
    print("\n💬 Commentaire sur les corrélations:")
    print("   • Intra-variables: fortes corrélations entre variables économiques")
    print("     (emp.var.rate, euribor3m, nr.employed sont liées)")
    print("   • Inter-variables (avec la cible):")
    print("     → 'duration' corrélation positive forte (+0.41)")
    print("     → 'euribor3m' corrélation négative modérée (-0.31)")
    print("     → 'nr.employed' corrélation négative modérée (-0.30)")
    print("     → Les variables économiques sont importantes")

# %%
# ============================================================
# 8. SÉLECTION DE VARIABLES
# ============================================================
print("\n" + "=" * 60)
print("8. SÉLECTION DE VARIABLES")
print("=" * 60)

print("✅ Pourrait-on avoir recours à la sélection de variables ? OUI")
print("\nJustification:")
print("   1. Réduction de la dimensionnalité (20 → 10-12 variables)")
print("   2. Amélioration des performances du modèle")
print("   3. Réduction du sur-apprentissage (overfitting)")
print("   4. Meilleure interprétabilité")
print("   5. Temps d'entraînement réduit")
print("\nMéthodes de sélection recommandées:")
print("   • Basée sur corrélation avec la cible")
print("   • Importance des features (Random Forest)")
print("   • Sélection univariée (SelectKBest)")
print("   • RFE (Recursive Feature Elimination)")
print("\n🎯 Variables à conserver en priorité (basé sur corrélations):")
target_corr = corr_with_target[corr_with_target.index != 'y_encoded']
top5_pos = target_corr[target_corr > 0].head(3)
top5_neg = target_corr[target_corr < 0].head(3)
print(f"   Variables positives: {list(top5_pos.index)}")
print(f"   Variables négatives: {list(top5_neg.index)}")

# %%
# ============================================================
# 9. NUAGES DE POINTS AVEC DROITES DE RÉGRESSION
# ============================================================
print("\n" + "=" * 60)
print("9. NUAGES DE POINTS AVEC DROITES DE RÉGRESSION")
print("=" * 60)

from scipy import stats

# Sélection des variables les plus importantes
important_vars = ['duration', 'euribor3m', 'nr.employed', 'emp.var.rate', 'age', 'campaign']

# Création du graphique
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
axes = axes.ravel()

for idx, var in enumerate(important_vars):
    if idx < len(axes) and var in df.columns:
        # Préparation des données
        x = df[var].values
        y = df['y'].map({'no': 0, 'yes': 1}).values
        
        # Nettoyer les valeurs NaN ou inf
        mask = ~(np.isnan(x) | np.isnan(y) | np.isinf(x) | np.isinf(y))
        x_clean = x[mask]
        y_clean = y[mask]
        
        # Nuage de points
        axes[idx].scatter(x_clean, y_clean, alpha=0.3, 
                         c=y_clean, cmap='coolwarm', s=10, edgecolors='none')
        
        # Calcul de la régression linéaire
        if len(x_clean) > 1 and len(np.unique(x_clean)) > 1:
            slope, intercept, r_value, p_value, std_err = stats.linregress(x_clean, y_clean)
            
            # Tracer la droite de régression
            x_line = np.array([x_clean.min(), x_clean.max()])
            y_line = intercept + slope * x_line
            axes[idx].plot(x_line, y_line, 'k--', linewidth=2, 
                          label=f'y = {slope:.4f}x + {intercept:.4f}\nR² = {r_value**2:.4f}')
            
            # Ajouter les paramètres comme texte
            axes[idx].legend(loc='upper right', fontsize=8)
        
        axes[idx].set_title(f'{var} vs Souscription', fontsize=11, fontweight='bold')
        axes[idx].set_xlabel(var, fontsize=9)
        axes[idx].set_ylabel('Souscription (0=Non, 1=Oui)', fontsize=9)
        axes[idx].grid(True, alpha=0.3)

plt.suptitle("Nuages de points avec droites de régression", fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('../reports/figures/scatter_plots.png', dpi=100, bbox_inches='tight')
plt.show()

# %%
# ============================================================
# 10. COMMENTAIRES DES RÉSULTATS
# ============================================================
print("\n" + "=" * 60)
print("10. COMMENTAIRES DES RÉSULTATS")
print("=" * 60)

print("""
📊 Analyse des nuages de points:

1. duration (durée de l'appel):
   • Forte séparation entre les classes
   • Les clients qui souscrivent ont des appels plus longs
   • R² relativement élevé → bonne relation linéaire
   ⚠️ Attention: la durée n'est connue qu'après l'appel (biais potentiel)

2. euribor3m (taux Euribor 3 mois):
   • Relation inverse avec la souscription
   • Quand taux bas, plus de souscriptions
   • Variable macroéconomique importante

3. nr.employed (nombre d'employés):
   • Corrélation négative modérée
   • Reflet de la santé économique

4. emp.var.rate (taux de variation emploi):
   • Lié à euribor3m et nr.employed
   • Multicolinéarité entre variables économiques

5. age:
   • Pas de séparation nette
   • Distribution similaire pour les deux classes

6. campaign (nombre de contacts):
   • Légère tendance inverse
   • Trop de contacts = moins de souscriptions (harcèlement)

⚠️ Conclusion:
   • Les variables économique et la durée sont les plus discriminantes
   • Les régressions linéaires montrent des tendances mais pas de séparation parfaite
   → Nécessité de modèles plus complexes (arbres, forêts, etc.)
""")

# %%
# ============================================================
# 11. CONCLUSION DE L'ANALYSE
# ============================================================
print("\n" + "=" * 60)
print("11. CONCLUSION DE L'ANALYSE")
print("=" * 60)

print("""
✅ RÉSUMÉ DE L'ANALYSE:

┌─────────────────────────────────────────────────────────────┐
│  Métadonnées                                                 │
├─────────────────────────────────────────────────────────────┤
│  • Dataset: Bank Telemarketing (UCI)                        │
│  • {:,} instances, 20 features                     │
│  • 10 numériques, 10 catégorielles                          │
│  • Problème: Classification binaire                         │
│  • Défi: Déséquilibre des classes (11.3% de 'yes')          │
└─────────────────────────────────────────────────────────────┘

┌─────────────────────────────────────────────────────────────┐
│  Variables les plus importantes (basé sur corrélations)     │
├─────────────────────────────────────────────────────────────┤
│  • duration (durée appel) : +0.41                           │
│  • euribor3m (taux) : -0.31                                 │
│  • nr.employed (emploi) : -0.30                             │
│  • emp.var.rate (variation emploi) : -0.30                  │
│  • cons.conf.idx (confiance) : -0.13                        │
└─────────────────────────────────────────────────────────────┘

┌─────────────────────────────────────────────────────────────┐
│  Recommandations pour la modélisation                       │
├─────────────────────────────────────────────────────────────┤
│  1. Prétraitement: Encoder variables catégorielles          │
│  2. Normalisation: StandardScaler (pour NN et SVM)          │
│  3. Gestion déséquilibre:                                    │
│     - Utiliser F1-Score comme métrique principale           │
│     - Stratified CV pour garder proportion                  │
│  4. Sélection variables: Top 10-12 plus importantes          │
│  5. Modèles recommandés:                                     │
│     • Random Forest (robuste, gère bien déséquilibre)       │
│     • Gradient Boosting (performances élevées)              │
│     • Réseaux de neurones (si assez de données)             │
└─────────────────────────────────────────────────────────────┘
""".format(df.shape[0]))

print("\n" + "=" * 60)
print("✅ NOTEBOOK 01 TERMINÉ - PASSER AU NOTEBOOK 02")
print("=" * 60)
```
